```python
import pandas as pd

dir_in_train = [f"C:/Users/hoang/Documents/Dataset_KLTN/ciciot2023_extracted/merge-processed/Incremental_1.3/session{i}_train.parquet" for i in range(0, 3)]
dir_in_test = [f"C:/Users/hoang/Documents/Dataset_KLTN/ciciot2023_extracted/merge-processed/Incremental_1.3/session{i}_test.parquet" for i in range(0, 3)]

all_rows = []

all_labels = set()

# Quét tất cả file train + test để lấy tất cả nhãn
for train_path in dir_in_train:
    test_path = train_path.replace("_train.parquet", "_test.parquet")

    df_train = pd.read_parquet(train_path)
    df_test = pd.read_parquet(test_path)

    all_labels.update(df_train["Label"].unique())
    all_labels.update(df_test["Label"].unique())

all_labels = sorted(all_labels)  # sắp xếp alphabet cho dễ nhìn

rows = []

df1 = pd.read_parquet(dir_in_train[2])
keys= df1['Label'].value_counts()
labels =list(keys.keys())

print(labels)
keys = {0: 'Benign', 1:'DDoS', 2: 'DoS', 3: 'Reconnaisance', 4: 'Vulnerability_Scan'}


for train_path in dir_in_train:
    name = train_path.split("/")[-1].replace("_train.parquet", "")
    test_path = train_path.replace("_train.parquet", "_test.parquet")

    df_train = pd.read_parquet(train_path)
    df_test = pd.read_parquet(test_path)

    train_counts = df_train["Label"].value_counts()
    test_counts = df_test["Label"].value_counts()

    row = {"File": name}

    # Nhóm Train
    for lbl in all_labels:
        row[f"Train: {keys[lbl]}"] = train_counts.get(lbl, 0)

    # Nhóm Test
    for lbl in all_labels:        
        row[f"Test: {keys[lbl]}"] = test_counts.get(lbl, 0)

    rows.append(row)

df = pd.DataFrame(rows)
print(df)
print(df.to_markdown(index=False))
```
